# Entrenamiento de modelos para rentabilidad y escenarios de campaña de palta

Este notebook construye datasets enriquecidos desde los CSV fuente, entrena modelos para apoyar S-1 y S-2, guarda artefactos en `models/` y exporta tablas listas para el dashboard en `outputs/`.

```python
from pathlib import Path
import json
import sys
import pandas as pd
ROOT = Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))
from modeling_pipeline import build_enriched_cost_dataset, build_price_features, train_models
pd.set_option('display.max_columns', 120)
pd.set_option('display.width', 180)
ROOT
```

## 1. Construcción de datasets enriquecidos

```python
cost_df = build_enriched_cost_dataset(ROOT)
price_df = build_price_features(ROOT)
print('Filas costo enriquecido:', len(cost_df))
print('Filas precio FOB:', len(price_df))
display(cost_df.head())
display(price_df.head())
```

## 2. Variables principales disponibles

```python
cost_summary = cost_df.groupby('ANIO').agg(
    registros=('costo_por_kg', 'count'),
    costo_mediano_kg=('costo_por_kg', 'median'),
    costo_promedio_kg=('costo_por_kg', 'mean'),
    produccion_total_kg=('produccion_kg', 'sum'),
    rendimiento_mediano_kg_ha=('rendimiento_kg_ha', 'median'),
    pct_outliers=('es_outlier', 'mean'),
).round(3)
display(cost_summary)
price_summary = price_df.groupby('destino').agg(
    registros=('precio_fob_por_kilogramo', 'count'),
    volumen_total=('volumen_exportado', 'sum'),
    precio_promedio=('precio_fob_por_kilogramo', 'mean'),
).sort_values('volumen_total', ascending=False).head(12).round(3)
display(price_summary)
```

## 3. Entrenamiento y evaluación

```python
result = train_models(ROOT, ROOT / 'models')
metrics = result['metrics']
metadata = result['metadata']
print(json.dumps(metrics, indent=2))
print('\nModelos guardados en:', ROOT / 'models')
```

## 4. Lectura de métricas

- `cost_model`: estima costo productivo por kg a partir de variables productivas, agronómicas y comerciales. Se entrena sin outliers IQR.
- `price_model`: estima precio FOB por kg según destino, semana, temporada, volumen, operaciones y señales recientes.
- `risk_model`: estima probabilidad de producción afectada cuando la variable `P223A` está disponible.

```python
for name, values in metrics.items():
    print(f'\n{name}')
    print(values)
print('\nArchivos generados:')
for path in sorted((ROOT / 'models').glob('*')):
    print('-', path.name)
for path in sorted((ROOT / 'outputs').glob('*.csv')):
    print('-', path.name)
```
